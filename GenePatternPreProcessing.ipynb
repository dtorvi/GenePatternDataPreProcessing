```python
## Question 1:  Function that Outputs GCT File as Pandas DataFrame along with summary
import pandas as pd
import matplotlib
def GCTDataFrame(pathToGCF, summary=False):
    df = pd.read_csv(pathToGCF, sep='\t',skiprows=2)
    print(df)
    if summary == True:
        print("Number of Rows:", df.shape[0])
        print("Number of Columns:", df.shape[1])
    return df
```

```python
## Question 2B
df1 = GCTDataFrame("BRCA_large_20783x40.gct")
#Question 3: Plot Histogram for specified columns
df2 = df1[["A7-A0DB-normal","A7-A13E-normal","BH-A0B3-primary","BH-A0B5-primary"]]
df2.hist()
```

```python
## Question 4: Using the data in the DataFrame from #2b, print out a DataFrame containing one row for each sample
columns = [column for column in df1]
means = []
medians = []
stdevs = []

for i in df1.columns:
    temp = df1[i]
    if temp.dtype =='float64':
        mean = temp.mean()
        median = temp.median()
        st = temp.std()
        means.append(mean)
        medians.append(median)
        stdevs.append(st)
```

```python
## Question 4 Continued
columns = df1.columns
newDF = pd.DataFrame()
columns = columns[2:]
print(len(columns))
newDF['Sample Name'] = columns
newDF['Mean'] = means
newDF['Medians'] = medians
newDF['StandardDeviation'] = stdevs
```

```python
##Question 5A: Using the data in the DataFrame from #2b, filter out (remove) the 10% most variable rows based on their standard deviations
filterNum = int(df1.shape[0] * 0.1)
stds = []
tempdf = pd.DataFrame()
for i in range(len(df1['Name'])):
    t = df1.iloc[i][2:]
    if t[1].dtype =='float64':
        s = t.std()
        stds.append(s)
tempdf["stdev"]= stds
f = tempdf["stdev"].sort_values(ascending = False)
indices = f.index[0:filterNum]
filteredDF = df1.drop(labels = indices, axis = 0)
filteredDF
```
